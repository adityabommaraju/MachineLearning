```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv('Coffee_Stores_Data.csv')
```

```python
df.head()
```

```python
df_18 = df[df['StoreID']==18]
```

```python
df_18.head()
```

```python
len(df_18['PLU'].unique())
```

```python
df_18.groupby(['PLU']).count().sort_values(by='StoreID',ascending=False)
```

```python
df_18 = df_18[df_18['PLU']==3000162]
```

```python
df_18.shape
```

```python
df_18['BusinessDate'] = pd.to_datetime(df_18['BusinessDate'])
```

```python
df_18.set_index('BusinessDate', inplace=True)
```

```python
df_18 = df_18['SoldQuantity']
```

```python
df_18 = pd.DataFrame(df_18)
```

```python
df_18.head()
```

```python
### Check whether data is stationary
from statsmodels.tsa.stattools import adfuller

adf, pvalue, usedlag_, nobs_, critical_values_, icbest_ = adfuller(df_18)
print("pvalue = ", pvalue, " if above 0.05, data is not stationary")
```

```python
df_18['year'] = [d.year for d in df_18.index]
df_18['month'] = [d.strftime('%b') for d in df_18.index]
years = df_18['year'].unique()
```

```python
df_18.head()
```

```python
sns.boxplot(x='year', y='SoldQuantity', data=df_18)
```

```python
sns.boxplot(x='month', y='SoldQuantity', data=df_18)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
decomposed = seasonal_decompose(df_18['SoldQuantity'],  
                            model ='additive')
```

```python
trend = decomposed.trend
seasonal = decomposed.seasonal
residual = decomposed.resid
```

```python
plt.figure(figsize=(16,8))
plt.subplot(411)
plt.plot(df_18['SoldQuantity'], label='Original', color='yellow')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(trend, label='Trend', color='yellow')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(seasonal, label='Seasonal', color='yellow')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(residual, label='Residual', color='yellow')
plt.legend(loc='upper left')
plt.show()
```

```python
from pmdarima.arima import auto_arima
```

```python
arima_model = auto_arima(df_18['SoldQuantity'], start_p = 1, d=1, start_q = 1, 
                          max_p = 5, max_q = 5, max_d=5, m = 12, 
                          start_P = 0, D=1, start_Q=0, max_P=5, max_D=5, max_Q=5,
                          seasonal = True, 
                          trace = True, 
                          error_action ='ignore',   
                          suppress_warnings = True,  
                          stepwise = True, n_fits=50)
```

```python
# SARIMAX(5, 1, 1)x(5, 1, [], 12
```

```python
size = int(len(df_18) * 0.80)
X_train, X_test = df_18[0:size], df_18[size:len(df_18)]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX 
  
model = SARIMAX(X_train['SoldQuantity'],  
                order = (5, 1, 1),  
                seasonal_order =(5, 1, 1, 12)) 
  
result = model.fit() 
result.summary() 
```

```python
start_index = 0
end_index = len(X_train)-1
train_prediction = result.predict(start_index, end_index)
```

```python
start_index = len(X_train)
end_index = len(df_18)-1
prediction = result.predict(start_index, end_index).rename('Predicted Sales') 
```
